## Helper Functions

Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img):
    """Applies the Canny transform"""
    
    th, bw = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
    return cv2.Canny(img, th/2, th)
    #return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image


def avg_lines_by_slope(lines, expected_left_slope = None, expected_right_slope = None, tolerance = None):
    left_slope = 0
    left_intercept= 0
    left_len = 0
    
    right_slope = 0
    right_intercept = 0
    right_len = 0
    
    for line in lines:
        for x1,y1,x2,y2 in line:
            delta_x = x2 - x1
            delta_y = y2 - y1
            if delta_x != 0:
                m = delta_y/delta_x
                # If the slope isn't near-horizontal, use it
                if abs(m) > .35:
                    b = y1 - m * x1
                    line_len = math.sqrt((y2-y1)**2 + delta_x ** 2)
                    if(m > 0):
                        # possitive slope, assume left
                        if expected_left_slope is None or abs(expected_left_slope - m) < tolerance:
                            # weighted average, length is the weight
                            left_slope = left_slope + line_len * m
                            left_intercept = left_intercept + line_len * b
                            left_len = left_len + line_len
                    else:
                        # negative slope, assume right
                        if expected_right_slope is None or abs(expected_right_slope - m) < tolerance:
                            # weighted average, length is the weight
                            right_slope = right_slope + line_len * m
                            right_intercept = right_intercept + line_len * b
                            right_len = right_len + line_len
    
    if left_len != 0:
        left_intercept = left_intercept / left_len
        left_slope = left_slope / left_len
    else:
        left_intercept = None
        left_slope = None
        
    if right_len != 0:
        right_intercept = right_intercept / right_len
        right_slope = right_slope / right_len
    else:
        right_intercept = None
        right_slope = None
    
    return [(left_slope, left_intercept), (right_slope, right_intercept)]



def lane_lines(lines, min_y, max_y):
    
    # first pass to get estimate of lane lines
    expected_line_avgs = avg_lines_by_slope(lines)
    
    # second pass, ignore lines too far off estimate
    line_avgs = avg_lines_by_slope(lines, expected_line_avgs[0][0], expected_line_avgs[1][0], 0.1)
    
    m = expected_line_avgs[0][0]
    b = expected_line_avgs[0][1]
    if line_avgs[0][0] is not None:
        m = line_avgs[0][0]
        b = line_avgs[0][1]
    
    left_line = [0,0,0,0]
    if m is not None:
        x1 = int((max_y - 1 - b)/m)
        x2 = int((min_y - b)/m)
        left_line = [x1, max_y, x2, min_y]
    
    m = expected_line_avgs[1][0]
    b = expected_line_avgs[1][1]
    if line_avgs[1][0] is not None:
        m = line_avgs[1][0]
        b = line_avgs[1][1]
        
    right_line = [0,0,0,0]
    if m is not None:
        x1 = int((max_y - 1 - b)/m)
        x2 = int((min_y - b)/m)
        right_line = [x1, max_y, x2, min_y]
    
    return [[left_line], [right_line]]


def draw_lines(img, lines, color=[255, 0, 0], thickness=2):
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    for line in lines:
        for x1,y1,x2,y2 in line:
            cv2.line(img, (x1, y1), (x2, y2), color, thickness)
    return line_img
    

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    return lines
    

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, α=0.8, β=1., λ=0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img * α + img * β + λ
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, α, img, β, λ)
````

Try tuning the various parameters, especially the low and high Canny thresholds as well as the Hough lines parameters.

````python
def gen_roi_vertices(image):
    """Get Trapazoidal Region Of Interest"""
    
    imshape = image.shape
    
    # Using bottom 40% of the image
    roi_percent_height = 0.40
    
    # top of ROI trapazoid will be 12% image width
    roi_percent_top_width = 0.12
    
    # bottom of ROI trapazoid will be 95% image width
    roi_percent_bottom_width = 0.95
    
    # trapazoid wil be offset 10 pixels from the bottom
    roi_bottom_offset = 10
    
    roi_y_top = int((1.0 - roi_percent_height) * imshape[0])
    roi_y_bottom = imshape[0] - roi_bottom_offset
    
    roi_x_left_bottom = int((imshape[1] * (1.0 - roi_percent_bottom_width))/2)
    roi_x_left_top = int((imshape[1] * (1.0 - roi_percent_top_width))/2)
    roi_x_right_bottom = int((imshape[1] * (1.0 + roi_percent_bottom_width))/2)
    roi_x_right_top = int((imshape[1] * (1.0 + roi_percent_top_width))/2)
    
    
    roi_left_m = (roi_x_left_top - roi_x_left_bottom) / (roi_y_top - roi_y_bottom)
    roi_left_b = roi_y_top - roi_left_m * roi_x_left_top
    
    roi_right_m = (roi_x_right_top - roi_x_right_bottom) / (roi_y_top - roi_y_bottom)
    roi_right_b = roi_y_top - roi_right_m * roi_x_right_top
    
    
    vertices = np.array([[(roi_x_left_bottom, roi_y_bottom),(roi_x_left_top, roi_y_top),
                          (roi_x_right_top, roi_y_top), (roi_x_right_bottom, roi_y_bottom)]],
                        dtype=np.int32)
    
    
    return vertices, roi_y_top


def detect_lane(image):
    """Detect left/right lane in image and return new image with the lanes highlighted.
    """
    imshape = image.shape
    
    # Stage 1, convert to gray scale
    gray = grayscale(image)
    
    # Stage 2, gaussian blur to smooth out image
    kernel_size = 5
    blur_gray = gaussian_blur(gray,kernel_size)
    
    # Stage 3, canny edge detection
    edges = canny(blur_gray)
    
    #Stage 4, generate ROI and remove edges outside it
    roi, roi_y_top = gen_roi_vertices(image)
    masked_edges = region_of_interest(edges, roi)
    
    #Stage 5, Hough thransforms to find lines
    
    rho = 5 # distance resolution in pixels of the Hough grid
    theta = np.pi/180 # angular resolution in radians of the Hough grid
    threshold = 6     # minimum number of votes (intersections in Hough grid cell)
    min_line_length = 10 #minimum number of pixels making up a line
    max_line_gap = 2    # maximum gap in pixels between connectable line segments
    line_image = np.copy(image)*0 # creating a blank to draw lines on
    
    lines = hough_lines(masked_edges, rho, theta, threshold, min_line_length, max_line_gap)
    
    # Stage 6, Covert the hough lines into "left lane" and "right lane" lines
    lines = lane_lines(lines, roi_y_top, imshape[0])
    
    # Make an image with just the lane lines
    lines_img = draw_lines(image, lines, [250,0,0], 10)
    
    # Overlay the lane line image on the original image
    lines_edges = weighted_img(lines_img, image, 0.5, 1.0, 0)
    return lines_edges

    
img_file_names = os.listdir("test_images/")

for img_file_name in img_file_names:
    image = mpimg.imread('test_images/' + img_file_name)
    lanes = detect_lane(image)
    plt.figure()
    plt.imshow(lanes)
````
